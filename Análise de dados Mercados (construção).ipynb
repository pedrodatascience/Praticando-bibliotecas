# ANÁLISE DOS DADOS

```python
#Nome das colunas
list(df_loja)
```

# RENOMEAÇÃO DAS COLUNAS

```python
#Utilização da função "rename" para renomear as colunas
df_loja = df_loja.rename(
columns= {"Store ID " : "ID_LOJA",
         "Store_Area" : "AREA_LOJA",
         "Items_Available" : "ITENS_VENDA",
         "Daily_Customer_Count" : "VISITAS_CLIENTES",
         "Store_Sales" : "VENDAS_LOJA"}
)
```

```python
#Variavéis após a renomeação
list(df_loja)
```

# CONVERSÃO DE JARDAS QUADRADAS PARA METROS QUADRADOS.

```python
# Conversão de Jardas Quadradas em Metros Quadrados
# 1 Jarda Quadrada = 0.86 Metro Quadrado
df_loja["AREA_LOJA"] = df_loja["AREA_LOJA"]*0.83
```
